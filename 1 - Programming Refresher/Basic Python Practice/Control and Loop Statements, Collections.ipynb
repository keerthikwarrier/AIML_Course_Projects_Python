```python
#If statement
x,y=23,45
if(x<y):
    print("x is less than y")
```

```python
#If Else statement
if(y<x):
    print("x is greater than y")
else:
    print("x is less than or equal to y")
```

```python
#If Elif Else statement
if(y<x):
    print("x is greater than y")
elif(x<y):
    print("x is less than y")
else:
    print("x is equal to y")
```

```python
#Nested If statement
if(y>=x):
    if(y>x):
        print("x is less than y")
    else:
        print("x is greater than y")
```

**FOR loop**
- i in the for loop as the local iterable variable.
- the number of for loop iterations =no of values generated by the range function/present in the collection variable


```python
for i in range(1,6):
    print(i)
```

**Combine for loop and if condition**

```python
for i in range(1,6):
    if i%2==0:
        print(i,"is an even number")
    else:
        print(i, "is not odd number")
```

**While Loop**
- while loop will execute only if the condition is met

```python
num=1
while num<11:
    print(num)
    num+=1
    
```

**Loop control statements**
- Break - exits the loop abruptly
- continue - it will skip that loop and continur with next

```python
#break
for i in "Good Morning":
    if i==' ':
        break
    print(i)
print("End of loop")
```

```python
#continue
for i in "Good Morning":
    if i==' ':
        continue
    print(i)
print("End of loop")
```

**Collection Data Types**
- Lists
  - Lists should be declared with a set of square braces only.
  - Lists are 1 dimensional and they are mutable(i.e) the values can be modified
  - Lists are ordered collection of items
  - Lists can contain heterogeneous data
- Tuples
  - Tuples should be declared with a set of round braces only.
  - Tuples are 1 dimensional and they are immutable(i.e) the values cannot be modified
  - Tuples are ordered collection of items
  - Tuples can contain heterogeneous data
- Set
  - Sets should be declared with a set of curly braces only.
  - Tuples are 1 dimensional and they are immutable(i.e) the values cannot be modified
  - Sets contain only unique values
  - Sets are unordered collection of items
  - Tuples can contain heterogeneous data
- Dictionaries
  - Dictionaries are declared with curly braces
  - Dictionaries are used to store the data in the form of key-value pairs (Keys are always supposed to be unique and maintain same datatype for keys)
  - Dictionaries are mutable
  

```python
#Lists - with square braces only
num_list=[1,2.0,3,'Keerthi',5,6]
print(num_list)

#indexing
num_list[3]

#Slicing - collection_variable_name[start index value:stop index value]
num_list=[1,2.0,3,'Keerthi','Hindi',5,6]
print(num_list[0:3])

#len() - returns the length of the list
print(len(num_list))

#Modifying
num_list[1]="Hari"
print(num_list)

num_list.append('Wonderful')
print(num_list)
```

```python
First_Tuple=(12,'Keerthi',4.5,'Hindi','Stupid','digit')
print(First_Tuple)
#type()
print(type(First_Tuple))

#Slicing
print(First_Tuple[0:3])

#indexing
print(First_Tuple[2])

#Modifying- You cannot modify a tuple as its immutable
First_Tuple[1]="Hari"
```

**How do you remove duplicate values in Python**
**Ans: Type cast the list to sets**


```python
num_list=[3,4,2,5,6,3]
#convert the list to set
num_set=set(num_list)
print(num_set)
set_to_list = list(num_set)
print(set_to_list)
```

```python
#Dictionaries
student_dict={101:['Keerthi','keerthi.k.warrier@gmail.com',29],
             102:['Hindus','fffgfg@gmail.com',30]
             }
#indexing
print(student_dict[101])
print(student_dict[101][2])

#slicing
print(student_dict[101][0:2])

#Delete a key
del student_dict[102]
print(student_dict)
```

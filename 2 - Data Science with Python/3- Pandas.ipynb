### Pandas
- It has 3 important function by its roots
  - Series - 1 dimesional data - A Pandas Series is like a column in a table. It is a one-dimensional array holding data of any type.
  - DataFrame - 2 dimensional data
  - Panels - 3 dimensional data

```python
import pandas as pd
```

#### Series 

```python
var=['Keerthi','Hari','Thakku']
Series_with_dimension=pd.Series(var)
Series_with_dimension
```

```python
var=['Keerthi','Hari','Thakku']
Series_with_dimension=pd.Series(var,index=["r1","r2","r3"])
Series_with_dimension
```

```python
#iloc will take the values based on the row or column value (startnng from 0) and not the index(r1,r2,r3). In this case, it will take the value which is in 1st row . 
Series_with_dimension.iloc[1]
```

```python
#This line uses where. the first argument checks the condition and prints the value if its true. The second argument is printed if the condition in first one is false
Series_with_dimension.where(Series_with_dimension.str.startswith('H'),'Doesn\'t start with H')
```

```python
var1=['Keerthi','Hari','Thakku']
dataFrame_with_dimension=pd.DataFrame(var1)
dataFrame_with_dimension
```

```python
import pandas as pd
date_range_data=pd.date_range(start='2024-01-01',end='2024-01-20',freq='D')
print(date_range_data)
```

```python
my_date={'my_Date':['2023-01-01','2023-01-20','2023-02-28']}
df=pd.DataFrame(my_date)
df

#Convert the data into datetime
df['my_Date']=pd.to_datetime(df['my_Date'])
df['my_Date'].dtype

```

```python
df['Day']=df['my_Date'].dt.day
df
```

```python
df['Month']=df['my_Date'].dt.month
df
```

```python
df['Year']=df['my_Date'].dt.year
df
```

```python
df=pd.DataFrame({'Date':pd.date_range(start='2023-01-01',periods=5)})
df['Weekday']=df['Date'].dt.weekday
df['IsWeekend']=df['Date'].dt.weekday // 5 == 1
print(df[['Date','Weekday','IsWeekend']])
```

```python
data={
    'Date':pd.date_range(start='2023-01-01',periods=10,freq='H'),
    'Value1':range(10),
    'Value2':range(10,20)
}
df=pd.DataFrame(data)
print(df)
delta=pd.Timedelta(days=3)
print(delta)
```

```python
df['Date']=pd.to_datetime(df['Date'])
df['FutureDate']=df['Date']+pd.Timedelta(weeks=2,days=3,hours=12)
print(df[['Date','FutureDate']])
```

```python
#counting occurances using DataFrame
df=pd.DataFrame({'Category':['A','B','V','C','C']})

category_counts=df['Category'].value_counts()
print(category_counts)
```

```python
#counting the occurance using collection
from collections import Counter
var=['A','B','V','C','C']
output=Counter(var)
output
```

```python
#counting the occurance using python Dictionary
count=0
var=['A','B','V','C','C']
output={}
for x in var:
    char_count=var.count(x)
    output[x]=char_count
output
```

```python
#if we have huge data we seperate it into chunks with 10 records (multiple dataframes)
df=pd.read_csv('Advertising.csv',chunksize=10)
for chunk in df:
    print(chunk)
```

```python
df=pd.read_csv('Advertising.csv')
#records with only TV > 200 is stored in df
df=df[df['TV']>200]
df
```

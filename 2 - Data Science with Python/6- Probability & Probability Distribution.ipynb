 ## Probability
- Probability is a mathematical term for the likelihood of any event happening.
-  A chance of any event occurring is a number between 0 and 1. 0 indicates the impossibility of the event and 1 indicates the certainity of the event occurring. 
- Probability anticipates risks and identify ways to manage such risks.
- Make forecasts about future events based on their likelihood
- Probability is powerful tool used to incorporate uncertainity in planning and decision making.
- Some of the places where probability is used are, weather forecasting, scientific research and healthcare.

### Random Variable in Probability
Random Variable is a variable that can take any value from a set of possible outcomes. Random Variable is different from other variables as it can take any values not just one.
For example:
- Lets say X is a random variable and we are tossing a coin. Now the Probaility of getting heads is 0.5. So, now X becomes 1 when we get head and X becomes 0 when we get tails.
- Similarly, Imagine we tossing the coin 10 times and we got heads all the 10 times. So, now X = 10 . If we get heads 9 times, then X=9

### Sum of Probabilities 
- **Disjoint Events** - Disjoint events are events where there is no chances of both events overlapping
  - Imagine there are 10 students in a school and 3 plays soccer and 4 plays basketball. What is the probaility of finding a student who plays either      soccer or basketball 
    P(Soccer)=3/10 , P(BasketBall)=4/10
   So, P(Soccer) + P(Basketball)=7/10

- **Joint Events**
  - For Joint Events - Joint events are events where there are chances of both events overlapping
  -  Example: If we take probaility of rain and wind occuring, there is a chance where both rain and wind occur together.

### Product Rule
- **Independent Events**
  - Independent Means one event is independent of the next onr. Example: Tossing the coin in second trail and getting heads or tails is not dependent on getting heads or tails in first trail.
  - **Product Rule in Independent Events**
    - P(Soccer and Room1) = P(Soccer).P(Room1) - Means Probability of students who play soccer and is in Room1
- **Dependent Events**
  - Dependent means the events are dependent to the other. Example: While playing chess, what happens in the first move is dependent on the 2nd and 10th move.

### Bernoullis Distribution
- To find the Probability for binary distribution( 0 and 1)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli, gaussian_kde

# Bernoulli distribution parameters
p = 0.6  # Probability of success

# Generating Bernoulli distribution data
data_bernoulli = bernoulli.rvs(p, size=1000)

# Plot histogram
plt.hist(data_bernoulli, bins=np.arange(-0.5, 2, 1), density=True, alpha=0.5, color='pink', edgecolor='black', linewidth=2, align='mid')

# Calculate the KDE
kde = gaussian_kde(data_bernoulli)
kde_xs = np.linspace(-1, 2, 300)
kde_ys = kde.pdf(kde_xs)

# Plot KDE
plt.plot(kde_xs, kde_ys, color='red')

# Add labels
plt.xlabel('Bernoulli')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
from scipy.stats import bernoulli
# Bernoulli distribution parameters
p = 0.6 

# Probability of success 
# Generating Bernoulli distribution data  (Out of 1000 data's, 60% will be success means 1)
#rvs- random variable sampling
data_bernoulli = bernoulli.rvs(p, size=1000) 
print(data_bernoulli)

# Unique values (0 and 1) and their probabilities #probability mass function 
x = [0, 1] 
#pmf shows how many or the mass of success percentage and failure percentage is generated
pmf_values = bernoulli.pmf(x, p) 
print(pmf_values)

# PMF for Bernoulli 
# Plot histogram (empirical data)
#from -0.5 to 2 with 1 interval. So, there will be 2 bins
plt.hist(data_bernoulli, bins=np.arange(-0.5, 2, 1), density=True, alpha=0.5, color='pink', edgecolor='black', linewidth=2, align='mid', label="Empirical Data")

# Plot PMF (theoretical probabilities) 
plt.scatter(x, pmf_values, color='red', marker='o', s=100, label="PMF (Theoretical)")

# Labels and legend plt.xlabel('Bernoulli Outcome') 
plt.ylabel('Probability') 
plt.xticks([0, 1]) 
plt.legend() 
plt.title("Bernoulli Distribution (p=0.6)") 

# Display the plot 
plt.show()
```

### Binomial Distribution
It gives success or failure , but outcome is multiple times. Example : tossing the coin multiple times gives multiple outcomes.
Binomial Distribution of a random variable is Probability of number of event occurances in number of bernoulli trails where, p is probability of event occurance and n is the number of trails

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters for the binomial distribution
n, p = 20, 0.8  # number of trials, probability of success per trial

# Generate binomially distributed data
data_binom = binom.rvs(n, p, size=10000)

# Plot histogram
plt.hist(data_binom, bins=range(n+2), density=True, alpha=0.5, color='green', label="Histogram")

# Calculate and plot PMF
x = np.arange(n+1)
pmf_values = binom.pmf(x, n, p)
plt.scatter(x, pmf_values, color='red', label="Binomial PMF", zorder=3)

# Add labels
plt.xlabel('Binomial')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

### Poisson Distribution
Poisson distribution describes the distribution of binary data from an infinite sample. 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Parameter for Poisson distribution
lambda_ = 10  # Mean number of events

# Generate Poisson-distributed data
data_poisson = poisson.rvs(lambda_, size=10000)

# Plot histogram
plt.hist(data_poisson, bins=range(min(data_poisson), max(data_poisson) + 2),
         density=True, alpha=0.5, color='blue', label="Histogram")

# Calculate and plot PMF
x = np.arange(min(data_poisson), max(data_poisson) + 1)
pmf_values = poisson.pmf(x, lambda_)
plt.scatter(x, pmf_values, color='red', label="Poisson PMF", zorder=3)

# Add labels and legend
plt.xlabel('Poisson')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import bernoulli, binom, poisson
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)


def bernoulli_example():
    """
    Example: Simulating fair coin flips (p=0.5)
    """
    p = 0.5  # probability of success (getting heads)
    n_trials = 1000

    # Generate random Bernoulli trials
    results = bernoulli.rvs(p, size=n_trials)

    # Calculate probabilities
    heads = np.sum(results)
    heads_prob = heads / n_trials

    print("\nBernoulli Distribution Example (Coin Flip)")
    print(f"Theoretical probability of heads: {p}")
    print(f"Observed probability of heads after {n_trials} flips: {heads_prob:.3f}")
    print(f"Number of heads: {heads}")
    print(f"Number of tails: {n_trials - heads}")


def binomial_example():
    """
    Example: Rolling a die 10 times and counting the number of sixes
    """
    n = 10  # number of trials
    p = 1 / 6  # probability of success (rolling a six)
    n_experiments = 1000

    # Generate random binomial trials
    results = binom.rvs(n, p, size=n_experiments)

    # Calculate mean number of successes
    mean_successes = np.mean(results)

    print("\nBinomial Distribution Example (Rolling Dice)")
    print(f"Theoretical mean number of sixes in {n} rolls: {n * p:.2f}")
    print(f"Observed mean number of sixes in {n_experiments} experiments: {mean_successes:.2f}")

    # Calculate probability of exactly 2 sixes
    prob_2 = binom.pmf(2, n, p)
    print(f"Probability of exactly 2 sixes in {n} rolls: {prob_2:.3f}")


def poisson_example():
    """
    Example: Customers arriving at a store
    Average of 5 customers per hour
    """
    lambda_param = 5  # average rate (customers per hour)
    hours = 100

    # Generate random Poisson arrivals
    arrivals = poisson.rvs(lambda_param, size=hours)

    # Calculate statistics
    mean_arrivals = np.mean(arrivals)

    print("\nPoisson Distribution Example (Customer Arrivals)")
    print(f"Expected number of customers per hour: {lambda_param}")
    print(f"Observed mean customers per hour over {hours} hours: {mean_arrivals:.2f}")

    # Calculate probability of exactly 3 customers in one hour
    prob_3 = poisson.pmf(3, lambda_param)
    print(f"Probability of exactly 3 customers in one hour: {prob_3:.3f}")

    # Plot distribution of arrivals
    plt.figure(figsize=(10, 6))
    plt.hist(arrivals, bins=range(max(arrivals) + 2), density=True, alpha=0.7, color='blue')
    plt.title('Distribution of Customer Arrivals per Hour')
    plt.xlabel('Number of Customers')
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)


# Run all examples
bernoulli_example()
binomial_example()
poisson_example()

# Show plots
plt.show()
```

### Central Limit Theorem
It should fall under a bell - curve(normal distribution)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create a non-normal population (exponential distribution)
np.random.seed(42)
population = np.random.exponential(scale=50, size=10000)  # Highly skewed data

# Step 2: Take multiple random samples & compute their means
sample_size = 30  # Each sample will have 30 data points
num_samples = 1000  # Number of samples we take

sample_means = [np.random.choice(population, sample_size).mean() for _ in range(num_samples)]

# Step 3: Convert to Pandas DataFrame
df = pd.DataFrame({"Sample Mean": sample_means})

# Step 4: Plot the histogram of sample means (CLT effect)
plt.figure(figsize=(8, 5))
sns.histplot(df["Sample Mean"], bins=30, kde=True, color="blue", alpha=0.7)
plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', label="Mean of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.legend()
plt.show()
```

### Estimation Theory

### Central Limit Theorem 
Process of almost making a uniform distribution into normal distribution by calculating its mean. This is called as Estimation Theory

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Step 1: Create Uniform data
np.random.seed(42)
population=np.random.uniform(low=0,high=100,size=10000)

#Step 2: Take many samples and compute the means
sample_size=30 #Each sample has 30 values
num_samples=1000 #Number of times we take samples

sample_means=[np.mean(np.random.choice(population,size=sample_size,replace=True)) for _ in range(num_samples)]

#Step 3: Plot original vs. saampling distribution
plt.figure(figsize=(12,5))

#Original Data Distribution (Uniform)
plt.subplot(1,2,1)
sns.histplot(population,bins=50,kde=True,color="blue")
plt.title("Original Population Distribution (Uniform)")


#Sampling Data Distribution (Normal)
plt.subplot(1,2,2)
sns.histplot(sample_means,bins=50,kde=True,color="red")
plt.title("Sampling Mean Data Distribution (Normal)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Step 1: Create a Skewed Population (Exponential Distribution)
np.random.seed(42)
population = np.random.exponential(scale=50, size=100000)  # Not normal!

# Step 2: Take Many Samples and Compute Sample Means
sample_size = 30  # Each sample has 30 values
num_samples = 1000  # Number of samples
sample_means = [np.mean(np.random.choice(population, size=sample_size, replace=True)) for _ in range(num_samples)]

# Step 3: Plot Population vs. Sampling Distribution
plt.figure(figsize=(12, 5))

# Original Data (Skewed)
plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color="blue")
plt.title("Original Population Distribution (Skewed)")

# Sample Means (Normal due to CLT)
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=50, kde=True, color="red")
plt.title("Sampling Distribution of the Mean (Normal by CLT)")

plt.show()

# Step 4: Estimate Population Mean Using Confidence Intervals
sample_mean = np.mean(sample_means)
sample_std = np.std(sample_means, ddof=1)  # Standard error
confidence_level = 0.95
z_value = norm.ppf((1 + confidence_level) / 2)  # Z-score for 95% confidence

margin_of_error = z_value * sample_std #lesser the value better the estimation
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Estimated Population Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")
```

### Bayes Theorem
It is combining historical data ans current data to do same CLT

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 📌 Step 1: Define Prior (Historical Data)
# Suppose previous elections had ~48% support for Candidate A.
# We model it as Beta(48, 52) assuming past election data was similar.
prior_alpha, prior_beta = 48, 52

# 📌 Step 2: Collect New Poll Data (Exit Poll)
poll_sample_size = 1000
poll_support = 550  # 55% of 1000 voters support Candidate A

# 📌 Step 3: Update Posterior Using Bayes' Theorem
posterior_alpha = prior_alpha + poll_support
posterior_beta = prior_beta + (poll_sample_size - poll_support)

# Plot Prior & Posterior Distributions
x = np.linspace(0, 1, 1000)
prior_dist = stats.beta.pdf(x, prior_alpha, prior_beta)
posterior_dist = stats.beta.pdf(x, posterior_alpha, posterior_beta)

plt.plot(x, prior_dist, label="Prior Belief (Historical Data)", linestyle='dashed')
plt.plot(x, posterior_dist, label="Updated Belief (Exit Poll Data)")
plt.xlabel("Proportion of Votes for Candidate A")
plt.ylabel("Density")
plt.legend()
plt.title("Bayesian Updating for Election Prediction")
plt.show()

# 📌 Step 4: Compute Bayesian Confidence Interval (95%)
lower_bound, upper_bound = stats.beta.interval(0.95, posterior_alpha, posterior_beta)
print(f"95% Bayesian Confidence Interval: ({lower_bound:.3f}, {upper_bound:.3f})")
```

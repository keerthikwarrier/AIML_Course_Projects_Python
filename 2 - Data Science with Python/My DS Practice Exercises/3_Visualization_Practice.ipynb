## __Steps to Perform:__
1. Create a line plot to visualize the trend of house prices over the years.


```python
import matplotlib.pyplot as plt

Housing_Data=pd.read_csv("housing_data.csv")
df=pd.DataFrame(Housing_Data)

plt.plot(df['SalePrice'])
plt.title("LinePlot: SalePrice")
plt.ylabel("Sale Price")
```

2. Use a scatter plot to visualize the relationship between two numerical variables, such as __LotArea__ and __SalePrice__.

```python
import matplotlib.pyplot as plt
Sqft=df['LotArea']
SalePrice=df['SalePrice']
plt.scatter(Sqft,SalePrice,c=Sqft,cmap="viridis")
plt.title("Scatter plot to compare Area with Price")
plt.xlabel("Area(Sqft)")
plt.ylabel("Price")

plt.legend()
plt.show()
```

3. Create a bar chart to show the count of houses in each __Neighborhood__.


```python
# Count the occurrences of each neighborhood
neighborhood_counts = df["Neighborhood"].value_counts()

# Create a bar chart
plt.figure(figsize=(12, 6))
neighborhood_counts.plot(kind="bar", color="skyblue", edgecolor="black")

# Formatting the plot
plt.xlabel("Neighborhood")
plt.ylabel("Number of Houses")
plt.title("Count of Houses in Each Neighborhood")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability

# Show the plot
plt.show()
```

4. Use a box plot to visualize the distribution of __SalePrice__ in each __Neighborhood__.


```python
# Create a box plot using pandas
plt.figure(figsize=(16, 6))
df.boxplot(column="SalePrice", by="Neighborhood", grid=False, vert=True, figsize=(16, 6), patch_artist=True)

# Formatting the plot
plt.xlabel("Neighborhood")
plt.ylabel("Sale Price")
plt.title("Distribution of SalePrice in Each Neighborhood")
plt.suptitle("")  # Remove default Pandas title
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability

# Show the plot
plt.show()
```

5. Create a pie chart to visualize the proportion of houses that fall into each __MSZoning__ category.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#read the csv file into a pandas dataframe and parsing dates
df=pd.read_csv("housing_data.csv",parse_dates=True)

df['MSZoning'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each zoning category
zoning_counts = df["MSZoning"].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(zoning_counts, labels=zoning_counts.index, autopct="%1.1f%%", startangle=140, colors=plt.cm.Paired.colors)
plt.title("Proportion of Houses in Each MSZoning Category")
plt.show()
```

6. Use a 3D scatter plot to visualize __LotArea__, __OverallQual__, and __SalePrice__ together.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#read the csv file into a pandas dataframe and parsing dates
df=pd.read_csv("housing_data.csv",parse_dates=True)

df.columns
```

```python
# Create a 3D scatter plot
ax = plt.axes(projection='3d')
ax.scatter(df['LotArea'], df['OverallQual'], df['SalePrice'])

# Set labels for each axis
ax.set_xlabel('Area in sqft')
ax.set_ylabel('Overall_Quality')
ax.set_zlabel('SalePrice')

# Display the 3D scatter plot
plt.show()
```

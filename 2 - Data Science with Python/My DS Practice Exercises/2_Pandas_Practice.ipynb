Create a detailed report on the monthly weather data by performing text manipulation, data sorting, and visualization to analyze temperature and precipitation trends.

__Data:__
The dataset contains daily observations of temperature and precipitation over a month.

```python
import pandas as pd

# Read CSV and parse 'Date' column as datetime
df = pd.read_csv("daily_weather_data.csv", parse_dates=['Date'])

# Display first few rows
print(df.head())
```

## __Steps to Perform:__

1. Textual manipulation
- Convert 'Day' to a string format with appropriate suffixes (1st, 2nd, 3rd, and so on)


```python
def add_suffix(day):
    if 10 <= day % 100 <= 20:  # Handles 11th to 20th
        suffix = "th"
    else:
        suffix = {1: "st", 2: "nd", 3: "rd"}.get(day % 10, "th")
    return f"{day}{suffix}"

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract day number and apply suffix function
df['Day'] = df['Date'].dt.day.apply(add_suffix)

print(df)
```

- Classify 'Temperature' into categories (Low, Medium, High) based on predefined thresholds

```python
def classify_temperature(temp):
    if temp < 15:
        return "Low"
    elif 15 <= temp <= 25:
        return "Medium"
    else:
        return "High"

# Apply classification function
df["Temperature Category"] = df["Temperature (°C)"].apply(classify_temperature)

print(df.head())
```

```python
#using Data Binning
#Define binedges and labels
segments=[-float("inf"), 15, 25, float("inf")]
category=['Low','Medium','High']
df['Temperature Category']=pd.cut(df['Temperature (°C)'],bins=segments,labels=category)

df[['Temperature (°C)','Temperature Category']]
```

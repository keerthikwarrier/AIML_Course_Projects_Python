```python
import yfinance as yf
import pandas as pd
from xgboost import XGBRegressor
import matplotlib.pyplot as plt

# Load 2024 data
ticker = "0P0000XV4Y.BO"
df = yf.download(ticker, start="2024-01-01", end="2024-12-31")
df = df[['Close']].rename(columns={'Close': 'NAV'})
df.index = pd.to_datetime(df.index)
```

```python
# Use last NAV of each month
monthly = df.resample('M').last()

# Add basic features
monthly['Return'] = monthly['NAV'].pct_change()
monthly['NAV_lag1'] = monthly['NAV'].shift(1)
monthly['Return_lag1'] = monthly['Return'].shift(1)
monthly['NAV_roll3'] = monthly['NAV'].rolling(3).mean().shift(1)
monthly['Return_roll3'] = monthly['Return'].rolling(3).mean().shift(1)
monthly = monthly.dropna().copy()
```

```python
X = monthly.drop(columns=['NAV'])
y = monthly['NAV']
```

```python
model = XGBRegressor(n_estimators=100, learning_rate=0.1)
model.fit(X, y)
```

```python
predictions = []
last_known = monthly.iloc[-1:].copy()

for month in range(1, 13):  # Jan to Dec 2025
    # Prepare input
    new_input = last_known.copy()
    #new_input['Month'] = month
    #new_input['Year'] = 2025

    pred_nav = model.predict(new_input.drop(columns=['NAV']))[0]
    predictions.append(pred_nav)

    # Update last_known for next prediction
    last_known['NAV'] = pred_nav
    last_known['NAV_lag1'] = pred_nav
    #last_known['Return_lag1'] = (pred_nav - new_input['NAV'].values[0]) / new_input['NAV'].values[0]
    last_known['NAV_roll3'] = (last_known['NAV_roll3'].values[0] * 2 + pred_nav) / 3
    #last_known['Return_roll3'] = (last_known['Return_roll3'].values[0] * 2 + last_known['Return_lag1']) / 3
```

```python
# Create time axis for 2025
future_dates = pd.date_range(start="2025-01-31", periods=12, freq='M')
pred_df = pd.DataFrame({'Predicted_NAV': predictions}, index=future_dates)

# Plot actual + predicted
plt.figure(figsize=(10, 5))
plt.plot(monthly.index, monthly['NAV'], label='Actual 2024 NAV', color='red')
plt.plot(pred_df.index, pred_df['Predicted_NAV'], label='Predicted 2025 NAV', color='green')
plt.title("JM Flexicap NAV Prediction for 2025")
plt.xlabel("Date")
plt.ylabel("NAV")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Which column is NAV here (The output parameter)
Column Name	Meaning

Date	    The date of the NAV entry

Price    	The NAV (Net Asset Value) per unit — this is the one you want ✅

Open / High / Low	Not relevant for NAV-based mutual funds (these are more relevant for ETFs traded on stock exchanges)

Vol (Volume)	Number of units traded — useful for liquidity analysis, not NAV

Change %	Day-over-day % change in NAV — useful to see volatility, but again, not NAV itself


```python
#Plot the NAV trend
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 6))
plt.plot(UTI_Gold_df.index, UTI_Gold_df['Price'], label='NAV')
plt.title('Historical NAV of UTI Gold ETF')
plt.xlabel('Date')
plt.ylabel('Price(NAV)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#To use the data in time series analysis we should convert the object columns into numerical 
vol_str=UTI_Gold_df['Vol.']
def convert_volume(vol_str):
    vol_str = str(vol_str).upper().strip()
    if 'K' in vol_str:
        return float(vol_str.replace('K', '')) * 1_000
    else:
         return pd.to_numeric(vol_str, errors='coerce')

UTI_Gold_df['Vol.'] = UTI_Gold_df['Vol.'].apply(convert_volume)
```

```python
UTI_Gold_df['Change %'] = UTI_Gold_df['Change %'].str.replace('%', '').astype(float)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from pandas.tseries.offsets import DateOffset
# Fit ARIMA model
model = ARIMA(UTI_Gold_df['Price'], order=(2, 1, 2),seasonal_order=(2, 1, 2, 12)) #p,d,q,s
model_fit = model.fit()
```

- p = 2 means the model uses two past values of the time series to predict the current value.
- d = 1 means the data needs to be differenced once to make it stationary.
- q = 2 means there are no moving average terms in the model.

```python
# Forecast next 180 days
forecast = model_fit.forecast(steps=180)
forecast_dates = [UTI_Gold_df.index[-1] + DateOffset(days=i) for i in range(1, 181)]
```

```python
# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(UTI_Gold_df.index, UTI_Gold_df['Price'], label='Historical NAV')
plt.plot(forecast_dates, forecast, color='green', label='Forecasted NAV')
plt.title('6-Month NAV Forecast - UTI Gold ETF FoF')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX 
model = SARIMAX(UTI_Gold_df['Price'],  
                order = (5, 0, 5),  
                seasonal_order =(5, 0, 5, 12)) 
model_fit = model.fit() 
```

```python
# Forecast next 180 days
forecast = model_fit.forecast(steps=180)
forecast_dates = [UTI_Gold_df.index[-1] + DateOffset(days=i) for i in range(1, 181)]
```

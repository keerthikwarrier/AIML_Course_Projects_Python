```python
#To use the data in time series analysis we should convert the object columns into numerical 
vol_str=UTI_Gold_df['ETF_Vol.']
def convert_volume(vol_str):
    vol_str = str(vol_str).upper().strip()
    if 'K' in vol_str:
        return float(vol_str.replace('K', '')) * 1_000
    else:
         return pd.to_numeric(vol_str, errors='coerce')

UTI_Gold_df['ETF_Vol.'] = UTI_Gold_df['ETF_Vol.'].apply(convert_volume)
```

```python
UTI_Gold_df['ETF_Change %'] = UTI_Gold_df['ETF_Change %'].str.replace('%', '').astype(float)
```

```python
UTI_Gold_df['USD_Change %'] = UTI_Gold_df['USD_Change %'].str.replace('%', '').astype(float)
```

```python
## Ensure the Time Series is Stationary
from statsmodels.tsa.stattools import adfuller

result = adfuller(UTI_Gold_df['MF_NAV'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

### Inferences
If the p-value is large (> 0.05), the series is non-stationary, and you should apply differencing (the d value).

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# ACF plot (for q)
plot_acf(UTI_Gold_df['MF_NAV'])
plt.show()

# PACF plot (for p)
plot_pacf(UTI_Gold_df['MF_NAV'])
plt.show()
```

## Prediction using SARIMAX model
__Step 1:__ To use SARIMAX, first we should have the future data of all the input variables. First, we have to get the input variables prediction values. This can be done using ARIMA within a for loop.

```python
from statsmodels.tsa.arima.model import ARIMA

forecast_steps = 15  # change as needed
exog_forecasts = {}

for col in X:  # exog_columns = list of your 15 column names
    model = ARIMA(UTI_Gold_df[col], order=(1,0,1))  # or use auto_arima for optimal params
    result = model.fit()
    forecast = result.forecast(steps=forecast_steps)
    exog_forecasts[col] = forecast.values
```

__Step 2:__ Create Dates for 15 forecast days and store the input variables predictions in the exogeneous variable forecasts values

```python
future_dates = pd.date_range(start=UTI_Gold_df.index[-1] + pd.Timedelta(days=1), periods=15, freq='B')  # 'B' for business days
X_exog = pd.DataFrame(exog_forecasts, index=future_dates)
X_exog
```

__Step 3:__ Now with these X_exog dataframe predict the future NAV by using SARIMAX model

```python
X_exog.shape
```

```python
X_exog
```

```python
y_endog=pd.DataFrame(UTI_Gold_df['MF_NAV'],index=future_dates)
y_endog
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Example SARIMA model for daily data with weekly seasonality (s = 7)
model = SARIMAX(y_endog,X_exog, order=(1,0,1),seasonal_order=(1,0,1,28))
model_fit = model.fit()
```

```python
# Forecast next 16 days
forecast_y= model_fit.forecast(steps=15,exog=y_exog)
forecast_y
```

```python
# Convert to DataFrame
forecast_y = pd.DataFrame({
    'Date': future_dates,
    'Forecast': forecast_y
})

# Set the date as index
forecast_y.set_index('Date', inplace=True)
# Final forecast DataFrame
print(forecast_y)
```

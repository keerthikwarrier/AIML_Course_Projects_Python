## UTI Gold ETF Fof Mutual Fund ( with MF, ETF, USD and Gold Data)

```python
import pandas as pd
UTI_Gold_df=pd.read_csv('UTI_Gold_ETF_MF_USD_01012023_31122024_ALL.csv',parse_dates=True,index_col='All_Date')
UTI_Gold_df.tail()
```

```python
UTI_Gold_df.index
```

```python
#Check for null values
UTI_Gold_df.isnull().sum()
```

```python
for column in UTI_Gold_df.columns:
    if UTI_Gold_df[column].isna().sum() > 0:
        UTI_Gold_df[column].bfill(inplace=True)
print(" Missing values after imputation:")
print(UTI_Gold_df.isna().sum())
```

```python
UTI_Gold_df.dtypes
```

```python
UTI_Gold_df.index = pd.to_datetime(UTI_Gold_df.index, errors='coerce')
```

```python
UTI_Gold_df.head()
```

```python
#Use a basic line plot
UTI_Gold_df.plot()
```

```python
#Plot the NAV trend
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 6))
plt.plot(UTI_Gold_df.index, UTI_Gold_df['MF_NAV'], label='NAV')
plt.title('Historical NAV of UTI Gold ETF')
plt.xlabel('Date')
plt.ylabel('Mutual_NAV')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#To use the data in time series analysis we should convert the object columns into numerical 
vol_str=UTI_Gold_df['ETF_Vol.']
def convert_volume(vol_str):
    vol_str = str(vol_str).upper().strip()
    if 'K' in vol_str:
        return float(vol_str.replace('K', '')) * 1_000
    else:
         return pd.to_numeric(vol_str, errors='coerce')

UTI_Gold_df['ETF_Vol.'] = UTI_Gold_df['ETF_Vol.'].apply(convert_volume)
```

```python
UTI_Gold_df['ETF_Change %'] = UTI_Gold_df['ETF_Change %'].str.replace('%', '').astype(float)
```

```python
UTI_Gold_df['USD_Change %'] = UTI_Gold_df['USD_Change %'].str.replace('%', '').astype(float)
```

```python
UTI_Gold_df.head()
```

```python
## Ensure the Time Series is Stationary
from statsmodels.tsa.stattools import adfuller

result = adfuller(UTI_Gold_df['MF_NAV'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

### Inferences
If the p-value is large (> 0.05), the series is non-stationary, and you should apply differencing (the d value).

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# ACF plot (for q)
plot_acf(UTI_Gold_df['MF_NAV'])
plt.show()

# PACF plot (for p)
plot_pacf(UTI_Gold_df['MF_NAV'])
plt.show()
```

```python
UTI_Gold_df.plot.box()
```

```python
y=UTI_Gold_df['MF_NAV']
X=UTI_Gold_df.drop('MF_NAV',axis=1)
```

```python
UTI_Gold_df.shape
```

```python
X.shape
```

```python
y.shape
```

```python
X.isnull().sum()
```

```python
for column in X.columns:
    if X[column].isna().sum() > 0:
        X[column].bfill(inplace=True)
print(" Missing values after imputation:")
print(X.isna().sum())
```

```python
from statsmodels.tsa.arima.model import ARIMA
from pandas.tseries.offsets import DateOffset
# Fit ARIMA model
model = ARIMA(UTI_Gold_df['MF_NAV'], order=(1, 1, 0), seasonal_order=(1, 1, 0, 28)) #p,d,q,s
model_fit = model.fit()
```

```python
# Forecast next 16 days
forecast = model_fit.forecast(steps=15)
forecast_dates = [UTI_Gold_df.index[-1] + DateOffset(days=i) for i in range(1, 16)]
```

```python
forecast_dates = pd.date_range(UTI_Gold_df.index[-1] + pd.Timedelta(days=1), periods=15, freq='B') 
```

```python
# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(UTI_Gold_df.index, UTI_Gold_df['MF_NAV'], label='Historical NAV')
plt.plot(forecast_dates, forecast, color='green', label='Forecasted NAV')
plt.title('15 days NAV Forecast - UTI Gold ETF FoF')
plt.xlabel('Date')
plt.ylabel('NAV')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Example SARIMA model for daily data with weekly seasonality (s = 7)
model = SARIMAX(UTI_Gold_df['MF_NAV'],X, order=(1,1,7),seasonal_order=(1,1,0,28))
model_fit = model.fit()
```

```python
# Forecast next 16 days
forecast = model_fit.forecast(steps=15)
forecast_dates = [UTI_Gold_df.index[-1] + DateOffset(days=i) for i in range(1, 16)]
```

```python
forecast_dates = pd.date_range(UTI_Gold_df.index[-1] + pd.Timedelta(days=1), periods=15, freq='B') 
```

```python
# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(UTI_Gold_df.index, UTI_Gold_df['MF_NAV'], label='Historical NAV')
plt.plot(forecast_dates, forecast, color='green', label='Forecasted NAV')
plt.title('15 days NAV Forecast - UTI Gold ETF FoF')
plt.xlabel('Date')
plt.ylabel('NAV')
plt.grid(True)
plt.legend()
plt.show()
```

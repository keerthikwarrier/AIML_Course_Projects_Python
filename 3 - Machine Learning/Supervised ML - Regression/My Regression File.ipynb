test_size is a parameter tht can accept any value between 0 and 1. But mind that 0 and 1 are exclude (i.e) it can accept any value>0 and <1

for ex, if test_size = 0.2, it implies that 20% of the data points are to be used for testing and the remaining 80% of the data points are to be used for training.

for ex, if test_size = 0.1, it implies that 10% of the data points are to be used for testing and the remaining 90% of the data points are to be used for training.

- test_size = 0 ==> None of the data points to be used for testing
- test_size = 1 ==> 100% of data to be used for testing  ==> this is the reason for the test_size parameter to never accept the values 0 and 1.

# Ridge Linear Regression or L2 regularisation

```python
#with default alpha value
from sklearn.linear_model import Ridge
ridge_model=Ridge()
ridge_model.fit(X_train,y_train)
y_train_pred_ridge=ridge_model.predict(X_train)
y_test_pred_ridge=ridge_model.predict(X_test)
print("RMSE of test data",np.sqrt(mean_squared_error(y_test,y_test_pred_ridge)))
print("RMSE of train data",np.sqrt(mean_squared_error(y_train,y_train_pred_ridge)))
```

```python
y_train_pred_low_alpha=ridge_mod_low_alpha.predict(X_train)
y_test_pred_low_alpha=ridge_mod_low_alpha.predict(X_test)
```

```python
print("RMSE of test data",np.sqrt(mean_squared_error(y_test,y_test_pred_low_alpha)))
```

```python
print("RMSE of train data",np.sqrt(mean_squared_error(y_train,y_train_pred_low_alpha)))
```

```python
ridge_cv_model = RidgeCV(alphas= np.random.randint(0, 1000, 100), cv=10).fit(X_train, y_train)

# Get the best alpha (regularization strength) selected by cross-validation
ridge_cv_model.alpha_
```

# ElasticNet Regularisation

```python
from sklearn.linear_model import ElasticNet
elastic_net_model=ElasticNet()
```

**alpha** : float, default=1.0
    Constant that multiplies the penalty terms. Defaults to 1.0.
    See the notes for the exact mathematical meaning of this
    parameter. ``alpha = 0`` is equivalent to an ordinary least square,
    solved by the :class:`LinearRegression` object. For numerical
    reasons, using ``alpha = 0`` with the ``Lasso`` object is not advised.
    Given this, you should use the :class:`LinearRegression` object.

**l1_ratio** : float, default=0.5
    The ElasticNet mixing parameter, with ``0 <= l1_ratio <= 1``. For
    ``l1_ratio = 0`` the penalty is an L2 penalty. ``For l1_ratio = 1`` it
    is an L1 penalty.  For ``0 < l1_ratio < 1``, the penalty is a
    combination of L1 and L2.

```python
elastic_net_model.fit(X_train,y_train)
```

```python
y_train_pred_ElasticNet=elastic_net_model.predict(X_train)
y_test_pred_ElasticNet=elastic_net_model.predict(X_test)
```

```python
print("RMSE of test data",np.sqrt(mean_squared_error(y_test,y_test_pred_ElasticNet)))
print("RMSE of train data",np.sqrt(mean_squared_error(y_train,y_train_pred_ElasticNet)))
```

# Polynomial Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv('position_salaries.csv')
dataset.head()
```

```python
dataset.info()
```

```python
# Let's identify the X and y
X=dataset.iloc[:,1:2].values # Level values
y=dataset.iloc[:,2].values # Salary values
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

```python
#Plot the model
def viz_linear():
    plt.scatter(X,y,color='red') 
    plt.plot(X,lin_reg.predict(X),color='blue')
    plt.title("Linear Regression model")
    plt.xlabel("Position")
    plt.ylabel("Salary")
    plt.show()
    return
```

```python
viz_linear()
```

#### NOTE: Now the problem is even if we fit the linear line but the datapoints are scattered and not fit with the line. So, try this in Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=5)
X_poly=poly_reg.fit_transform(X)
poly_reg=LinearRegression()
poly_reg.fit(X_poly,y)
```

```python
#Plot the model
def viz_polynomial():
    plt.scatter(X,y,color='red') 
    plt.plot(X,poly_reg.predict(X_poly),color='blue')
    plt.title("Linear Regression model with Polynomial degree 4")
    plt.xlabel("Position level")
    plt.ylabel("Salary")
    plt.show()
    return
```

```python
viz_polynomial()
```

```python
import pandas as pd
```

```python
ratings_df = pd.read_csv(r"D:\SimpliLearn\Machine+Learning\Recommender systems\ml-100k\u.data", sep = '\t',
                        names=['user_id', 'item_id', 'rating','timestamp'])
```

```python
ratings_df.head()
```

```python
movies_df = pd.read_csv(r"D:\SimpliLearn\Machine+Learning\Recommender systems\ml-100k\u.item", sep='|',
                       names=['movie_id','movie_title','release_date',
                            'video_release_date','IMDb URL','unknown',
                            'Action','Adventure','Animation','Children',
                            'Comedy','Crime','Documentary','Drama','Fantasy',
                            'Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi',
                            'Thriller','War','Western'],
                       encoding = "latin-1")
```

```python
movies_df.head()
```

```python
movies = movies_df.iloc[:,[0,1]]
movies.head(10)
```

```python
movies.shape
```

# Building the user-item matrix

```python
learning_Matrix= ratings_df.pivot_table(index=['item_id'], columns=['user_id'],
                      values='rating')
learning_Matrix.head()
```

```python
learning_Matrix.fillna(0, inplace=True)
learning_Matrix.tail(20)
```

# Finding out the cosine similarity between the items(movies)

```python
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
movie_similarity = pairwise_distances(learning_Matrix, metric = "cosine")
```

```python
pd.DataFrame(movie_similarity)
```

Based on our understanding from the above matrix, we fill the diagnol values with 0

```python
np.fill_diagonal(movie_similarity,0)

movie_similarity_df = pd.DataFrame(movie_similarity)
```

```python
movie_similarity_df
```

# Let's see the recommender system in action

```python
try:
    user_inp = input("Enter the reference movie title based on which recommendations are to be made:")
    movie_id = movies[movies["movie_title"] == user_inp].index.tolist()

    movie_id = movie_id[0] #extracted the movie id of the input given by the user

    movies["similarity"]= movie_similarity_df.iloc[movie_id]

    print("Recommended movies based on your choice of",user_inp,": \n",
     movies.sort_values(["similarity"], ascending= False)[1:10])
except:
    print("The movie title entered by you does not exist in the list. However, below are the top 10 recommended movies in general.")
    print(movies.head(10))
```

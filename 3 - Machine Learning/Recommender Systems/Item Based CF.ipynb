# Inferences
So in our datasets, there are 943 users, 1682 to movies and 100000 ratings provided.

## Building the User-Item Matrix

```python
User_Item_Matrix=ratings_df.pivot_table(index=['item_id'],columns=['user_id'],values='rating_by_user')
User_Item_Matrix
```

```python
#Fill the NaN values with 0 as we cannot work with null values and we cannot randomly give ratings using Imputing techniques
User_Item_Matrix.fillna(0,inplace=True)
User_Item_Matrix
```

## Finding the cosine similarity between the items (movies)

```python
from sklearn.metrics import pairwise_distances
import numpy as np
```

NOTE: When you use pairwise_distances(User_Item_Matrix, metric='cosine'), pairwise_distances treats each row as an individual item (or "sample"), and computes the distance between every pair of rows.

```python
movie_similarity=pairwise_distances(User_Item_Matrix,metric='cosine')
pd.DataFrame(movie_similarity)
```

```python
np.fill_diagonal(movie_similarity,0)
movie_similarity=pd.DataFrame(movie_similarity)
movie_similarity
```

## Let's see Recommender Systems

```python
try:
    user_inp = input("Enter the reference movie title based on which the recommendations are to be made: ")
    movie_id = movies[movies['movie_name_year'] == user_inp].index.tolist()
    movie_id = movie_id[0] #extracted the movie_id of the movie given by the user

    movies['Similarity']=movie_similarity.iloc[movie_id]

    print("Recommend movies based on your choice of",user_inp,":\n ",movies.sort_values(['Similarity'],ascending=False)[1:10])
except:
    print("The movie title entered by the user doesnot exist in the list. However, below are the top 10 recommended movies in general")
    print(movies.head(10))
```

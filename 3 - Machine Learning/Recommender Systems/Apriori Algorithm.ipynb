### Apriori Algorithm Parameters
1. Support - his metric measures how frequently an item appears in the dataset relative to the total number of transactions. A higher support indicates a more significant presence of the itemset in the dataset. Support tells us how often a particular item or combination of items appears in all the transactions (“Bread is bought in 20% of all transactions.”)

2. Confidence - likelihood of one product purchased when the other product is already bought. Confidence assesses the likelihood that an item Y is purchased when item X is purchased. It provides insight into the strength of the association between two items. Confidence tells us how often items go together. (“If bread is bought, butter is bought 75% of the time.”)

3. Lift - Lift evaluates how much more likely two items are to be purchased together compared to being purchased independently. A lift greater than 1 suggests a strong positive association. Lift shows how strong the connection is between items. (“Bread and butter are much more likely to be bought together than by chance.”)

### Apriori Algorithm can work only on lists


```python
#Convert a dataset into a list
records = []
for i in range(0,7501):
    records.append([str(store_df.values[i,j]) for j in range(0,20)])
print(records[0:3])
```

```python
from apyori import apriori
```

```python
#max_length is the number of products you want to bundle, min_lift is the strength of relation of the products
association_rules=apriori(records,min_lift=3,max_length=2,min_confidence=0.2,min_support=0.0045) 
association_rules_list=list(association_rules)
association_rules_list
```

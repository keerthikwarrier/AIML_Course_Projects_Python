One hot Enoding the labels

```python
y_train_one_hot=tf.keras.utils.to_categorical(y_train,10)
y_test_one_hot=tf.keras.utils.to_categorical(y_test,10)
```

**Before creating the model you should flatten the data . As it should be 1 dimensional . So we should reshape it from 28 by 28 to 784**

```python
x_train_flattened=x_train.reshape(60000,784)
x_test_flattened=x_test.reshape(10000,784)
print(x_train_flattened.shape)
print(x_test_flattened.shape)
```

## **So, the Preprocessing steps are completed**

#**Model Development**
**Sequential** - We are building a Sequential model. In this method the Layers( Input , hidden , Output) are stacked one after the other.

**Dense** - Dense is used build a fully connected Neural Network means every node in 1 layer is connected to each and every node in next layer (completely inter connected not intra connected)


```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

**Understand the number of nodes in each layer**

**Input Layer** - The number of nodes in input layer is same as number of features. There is no assumption in Deep Learning. The features can be independent or dependent to each other.

**Output Layer** - the number of nodes in output layer depends on the type of our data whether we use Regression or Classification models. If Regression we will have 1 node, Classification the number nodes is same as the number of classes in dataset. Number of nodes also depends on the activation function.

**Hidden Layer** - This is the part which the model builder can decide which its a hyperparameters. We can decide the number of hidden layers as well as the number of nodes in hidden layer.


## **Building the basic Neural Network model with only Input and Output Layers**

```python
#Lets begin with the Neural Network Model
mnist_digits_model=Sequential()
```

```python
mnist_digits_model.add(Dense(10,input_shape=(784,),activation='softmax'))
```

***The input layer will be created with your immediate subsequent layer it can be hidden or output layer***

Let's start building our model with only input and output layers without hidden layers.
So, here there will 10 nodes in Output Layer as our output has 10 classes 0 to 9 digits. So, the value in Dense function is how many nodes will be there in our immediate layer from input layer

**Activation function** is required for our output layer.Here I am using Softmax. Input doesn't use activation functions

```python
mnist_digits_model.summary()
```

Trainable params is the total number of weights and bias that are to be learnt as part of the training phase of this model.

Total number of weights + bias = 784 * 10 weights + 1 * 10 bias = 7850 Trainable Params

```python
#compile the model
mnist_digits_model.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
```

Here, We are using **SGD** as the **optimizer**,

**loss** can be decided based on whether you one hot encoded the data. Use **categorical_crossentropy** if you One hot encoded the data else use **sparse_categorical_crossentropy**

As We are creating a classification model we can use **evaluation metrics** as **accuracy**

```python
#Train the model
training=mnist_digits_model.fit(x_train_flattened,y_train_one_hot,epochs=5,validation_split=0.2)
```

**Epochs** is the number of complete pass/passes through our dataset. Here Epochs is 5 . So the model pass through the whole dataset 5 times.
For example, We have 60000 image samples and if our epoch is 1 and batch_size is 5000. Then, to complete 1 epoch, the training step would require 12 iterations as 60000/5000 =12
the no. of iterations = training_data_size/batch_size

```python
import matplotlib.pyplot as plt
%matplotlib inline
# list all data in training , training has a dictionary called history and its printing all the keys in history dictionary
print(training.history.keys())

# summarize training for accuracy
plt.plot(training.history['accuracy']) # training accuracy values
plt.plot(training.history['val_accuracy']) #validation accuracy values
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

# summarize traning for loss
plt.plot(training.history['loss']) # training loss values
plt.plot(training.history['val_loss']) #validation loss values
plt.title('model los')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
#Evaluate the model with test
test_loss,test_acc=mnist_digits_model.evaluate(x_test_flattened,y_test_one_hot)
print("Test Accuracy:",test_acc)
print("Test Loss:",test_loss)
```

### **Building the NN model with input, output and one hidden layer**

```python
mnist_digits_model_hidden=Sequential()
```

```python
mnist_digits_model_hidden.add(Dense(20,input_shape=(784,),activation='relu')) #input layer with hidden layer (20 nodes)
mnist_digits_model_hidden.add(Dense(10,activation='softmax')) #output layer
```

```python
mnist_digits_model_hidden.summary()
```

Here, the trainable params will be calculated as:

Input hidden, 20*784 weights+ 20 * 1 bias = 15700

Output layer, 10 * 20 weights + 10 * 1= 210
So, 15700 + 210 = 15910

---




```python
#compile the model
mnist_digits_model_hidden.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#train the model
training2=mnist_digits_model_hidden.fit(x_train_flattened,y_train_one_hot,epochs=5,validation_split=0.2)
```

```python
#Evaluate the model with test
test_loss,test_acc=mnist_digits_model_hidden.evaluate(x_test_flattened,y_test_one_hot)
print("Test Accuracy:",test_acc)
print("Test Loss:",test_loss)
```

## **Implementing Dropout**
Build the model with 2 hidden layer and 1 dropout layer. this model would have 4 layers in total - 1 input layer, 2 hidden layer and 1 output layer such that dropout layer will be applied to 1 hidden layer.

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model3=Sequential()
model3.add(Dense(40,name="first_hidden_layer",input_shape=(784,),activation='relu'))
model3.add(Dropout(0.2))
model3.add(Dense(20,name="second_hidden_layer",activation='relu'))
model3.add(Dense(10,name="output_layer",activation='softmax'))
```

```python
model3.summary()
```

```python
model3.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#Training the model
training3=model3.fit(x_train_flattened,y_train_one_hot,epochs=5,validation_split=0.2)
```

```python
#Evaluate the model with test
test_loss,test_acc=model3.evaluate(x_test_flattened,y_test_one_hot)
print("Test Accuracy:",test_acc)
print("Test Loss:",test_loss)
```

## **Implementing Batch Normalization**
Build the model with 2 hidden layer and 1 Batch Normalization layer. this model would have 4 layers in total - 1 input layer, 2 hidden layer and 1 output layer such that Batch Normalization layer will be applied to 1 hidden layer.

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,BatchNormalization
```

```python
model4=Sequential()
model4.add(Dense(50,name="first_hidden_layer",input_shape=(784,),activation='relu'))
model4.add(BatchNormalization()) #The first_hidden_layer will be batch normalized before sending it to second_hidden_layer
model4.add(Dense(20,name="second_hidden_layer",activation='relu'))
model4.add(Dense(10,name="output_layer",activation='softmax'))
```

```python
model4.summary()
```

```python
model4.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
training4=model4.fit(x_train_flattened,y_train_one_hot,epochs=5,validation_split=0.2)
```

```python
#Evaluate the model with test
test_loss,test_acc=model4.evaluate(x_test_flattened,y_test_one_hot)
print("Test Accuracy:",test_acc)
print("Test Loss:",test_loss)
```
